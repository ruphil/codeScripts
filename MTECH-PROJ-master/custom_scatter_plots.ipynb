```python
import numpy as np
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import scipy
```

```python
num_of_inputs = 100
outliers = 20
error_range = 10

x = np.random.uniform(0, 40, size=num_of_inputs)
y = np.array([])

for x_in in x:
    y = np.append(y, x_in + np.random.randint(-error_range, error_range))

for i in range(outliers):
    x = np.append(x, np.random.randint(0, 40))
    y = np.append(y, np.random.randint(0, 40))
    
# print(y.shape)
_, intercept, r_value, p_value, std_err = scipy.stats.linregress(x, y)
print("R_squared", r_value)
v = [0, 40, 0, 40]
plt.axis(v)
plt.xlabel("Y_Actual")
plt.ylabel("Y_Predicted")
plt.scatter(x, y)
plt.show()
```
